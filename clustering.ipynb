## Selecting Dimensionality Reduction and Hypertuning Parameters of Clustering Algorithms with Pipeline and GridSearchCV

```python
random_state = 42
scaler = StandardScaler()
data_std = scaler.fit_transform(all_individuals.drop(['Name', 'Team', 'Salary'], axis = 1))

pca = PCA(random_state = random_state )
pca.fit(data_std)

plt.figure(figsize = (10,8))
plt.plot(range(1,145), pca.explained_variance_ratio_.cumsum(), marker = 'o', linestyle = '--')
plt.title('Explaine Variance by Components')
plt.ylabel('Cumulative Explained Variance')
plt.xlabel('Number of Components')
```

```python
def silhouette_score(estimator, X):

    """Takes in pipeline estimator and dataset as arguments.
    
       Function makes predictions of dataset using estimator and computes silhouette score.
       
       Returns computed silhouette score."""
    
    estimator[0].fit(X)
    new = estimator[0].transform(X)
    estimator[1].fit(new)
    new = estimator[1].transform(new)
    clusters = estimator[2].fit_predict(new)
    score = metrics.silhouette_score(new, clusters)
    return score

def ch_criterion(estimator, X):

    """Takes in pipeline estimator and dataset as arguments.
    
       Function makes predictions of dataset using estimator and computes calinski-harabasz criterion.
       
       Returns computed calinski-harabasz criterion."""
 
    estimator[0].fit(X)
    new = estimator[0].transform(X)
    estimator[1].fit(new)
    new = estimator[1].transform(new)
    clusters = estimator[2].fit_predict(new)
    score = metrics.calinski_harabasz_score(new, clusters)
    return score


#drop unnecessary columns from dataset
X = all_individuals.drop(['Name', 'Team', 'Salary'], axis = 1)

#set random state
random_state = 42

#set cv to zero
cv = [(slice(None), slice(None))]

#create multi-scoring dictionary for gridsearch
scoring = {'Silhouette Score': silhouette_score, 'CH Criterion': ch_criterion}

#create algorithm dictionary for gridsearch
clusteralgs = [KMeans(random_state=random_state),
               MiniBatchKMeans(random_state=random_state),
               GaussianMixture(random_state=random_state),
               Birch()
              ]

#create parameters dictionary for gridsearch
params = [##KMeans
          {'pca__n_components': list(range(2,14)),
           'clust__n_clusters': list(range(15,56)),
           'clust__init': ['k-means++', 'random'],
           'clust__algorithm': ['full', 'elkan']}, 
          ##MiniBatchKMeans
          {'pca__n_components': list(range(2,14)),
           'clust__n_clusters': list(range(15,56)),
           'clust__init': ['k-means++', 'random'],
           'clust__reassignment_ratio': [0.01, 0.02, 0.03],
           'clust__batch_size': [50, 100, 150]},
          ##GaussianMixture 
          {'pca__n_components': list(range(2,14)),
           'clust__n_components': list(range(15,56)),
           'clust__covariance_type': ['full', 'tied', 'diag', 'spherical'],
           'clust__init_params': ['kmeans', 'random'],
           'clust__max_iter': [50, 75, 100]},
          ##Birch
          {'pca__n_components': list(range(2,14)),
           'clust__n_clusters': list(range(15,56)),
           'clust__branching_factor': [25, 45, 50, 55, 100],
           'clust__threshold':[0.3, 0.4, 0.5, 0.6]}
         ]

#create empty lists to keep best pipelines and best clfs
pipelines = []
best_clustalgs = []

#for loop iterating through each combination of classifier and its parameters
for clust, param in zip(clusteralgs, params):
    
    #set pipeline with list of standard scalar, pca and cluster algorithm
    for scaler in [StandardScaler(), MinMaxScaler()]:
        pipe = Pipeline([('ss', scaler),
                         ('pca', PCA(random_state=random_state)),
                         ('clust', clust)])

        #conduct gridsearch over parameter variable running on all backend workers
        gs = GridSearchCV(estimator=pipe,
                      param_grid=param,
                      scoring=scoring,
                      refit='Silhouette Score',
                      cv=cv,
                      verbose=1,
                      n_jobs=-1)

        #fit model
        gs.fit(X)
        #save results
        results = gs.cv_results_
        #append best estimator
        best_clustalgs.append(gs.best_estimator_)
        #append pipeline
        pipelines.append(pipe)
        #print cluster algorithm and its name
        print(clust.__class__.__name__, 'with', scaler.__class__.__name__)
        #print params of best model
        print('Best params:', gs.best_params_)
        #print silhouette score
        print('Silhouette Score:', silhouette_score(gs.best_estimator_, X))
        #print C-H Criterion
        print('C-H Criterion:', ch_criterion(gs.best_estimator_, X))
        
        #plot best estimators
        fig, (ax1, ax2) = plt.subplots(2, figsize=(10, 10))
        #set title
        fig.suptitle("GridSearchCV Evaluation Scores by Number of Clusters",
                  fontsize=16)
        #set subplot x and y labels
        ax1.set_xlabel("n_clusters")
        ax1.set_ylabel("Silhouette Score")
        #set subplot x and y labels
        ax2.set_xlabel("n_clusters")
        ax2.set_ylabel("C-H Criterion")
        
        #find index of best silhouette score for the gridsearch
        best_index = np.nonzero(results['rank_test_Silhouette Score'] == 1)[0][0]
        
        #if else statements to find number of clusters in best estimator
        if clust.__class__.__name__ == 'KMeans':
            num_of_clusters = results['param_clust__n_clusters'][best_index]
            div = 12
        elif clust.__class__.__name__ == 'MiniBatchKMeans':
            num_of_clusters = results['param_clust__n_clusters'][best_index]
            div = 36
        elif clust.__class__.__name__ == 'Birch':
            num_of_clusters = results['param_clust__n_clusters'][best_index]
            div = 48
        else:
            num_of_clusters = results['param_clust__n_components'][best_index]
            div = 12

        #for loop iterating through a range from 15 to 56, exclusive, to create lists for x-axis and y-axis
        #where x-axis is number of clusters and y-axis is corresponding silhouette score
        x_axis = []
        y_axis_ss = []
        for x in range(15,56):
            difference = num_of_clusters - x
            x_axis.append(x)
            y_axis_ss.append(results['mean_test_Silhouette Score'][best_index - int(difference * div)])

        #for loop iterating through a range from 15 to 56, exclusive, to create list for y-axis
        #where y-axis is corresponding c-h score
        y_axis_ch = []
        for x in range(15,56):
            difference = num_of_clusters - x
            y_axis_ch.append(results['mean_test_CH Criterion'][best_index - int(difference * div)])
        
        #
        ax1.set_ylim(min(y_axis_ss) - .05, max(y_axis_ss) + .05)
        ax1.plot(x_axis, y_axis_ss, linestyle = '--', color='g', label='Silhouette Score')
        best_score_ss = results['mean_test_Silhouette Score'][best_index]
        # Plot a dotted vertical line at the best score for that scorer marked by x
        ax1.plot([x_axis[num_of_clusters - 15],] * 2, [0, best_score_ss],
                linestyle='-.', color='k', marker='x', markeredgewidth=3, ms=8)
        # Annotate the best score for that scorer
        ax1.annotate("%0.2f" % best_score_ss, (x_axis[num_of_clusters - 15], best_score_ss + 0.005))
        
        ax2.set_ylim(min(y_axis_ch) - 10, max(y_axis_ch) + 10)
        ax2.plot(x_axis, y_axis_ch, linestyle = '--', color='g', label='CH Criterion')
        best_score_ch = results['mean_test_CH Criterion'][best_index]
        # Plot a dotted vertical line at the best score for that scorer marked by x
        ax2.plot([x_axis[num_of_clusters - 15],] * 2, [0, best_score_ch],
                linestyle='-.', color='k', marker='x', markeredgewidth=3, ms=8)
        # Annotate the best score for that scorer
        ax2.annotate("%0.2f" % best_score_ch, (x_axis[num_of_clusters - 15], best_score_ch + 0.005))

        ax1.legend(loc="best")
        ax1.grid(b = True, which = 'major')
        ax1.grid(b = True, which = 'minor')
        ax2.legend(loc="best")
        ax2.grid(b = True, which = 'major')
        ax2.grid(b = True, which = 'minor')
        plt.show()
```

## Making Classifier to Perform Feature Selection and Tune Model Parameters

```python
from sklearn.base import BaseEstimator
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn import preprocessing

class XGBoostWithEarlyStop(BaseEstimator):
    def __init__(self, early_stopping_rounds=50, test_size=0.30, 
                 eval_metric='mlogloss', **estimator_params):
        self.early_stopping_rounds = early_stopping_rounds
        self.test_size = test_size
        self.eval_metric=eval_metric     
        if self.estimator is not None:
            self.set_params(**estimator_params)

    def set_params(self, **params):
        return self.estimator.set_params(**params)

    def get_params(self, **params):
        return self.estimator.get_params()

    def fit(self, X, y):
        x_train, x_val, y_train, y_val = train_test_split(X, y, test_size=self.test_size, random_state = 40)
        self.estimator.fit(x_train, y_train, 
                           early_stopping_rounds=self.early_stopping_rounds, 
                           eval_metric=self.eval_metric, eval_set=[(x_val, y_val)])
        return self

    def predict(self, X):
        return self.estimator.predict(X)

class XGBoostClassifierWithEarlyStop(XGBoostWithEarlyStop):
    def __init__(self, *args, **kwargs):
        self.estimator = XGBClassifier()
        super(XGBoostClassifierWithEarlyStop, self).__init__(*args, **kwargs)
```

```python
def silhouette_score(estimator, X):

    """Takes in pipeline estimator and dataset as arguments.
    
       Function makes predictions of dataset using estimator and computes silhouette score.
       
       Returns computed silhouette score."""
    
    estimator[0].fit(X)
    new = estimator[0].transform(X)
    estimator[1].fit(new)
    new = estimator[1].transform(new)
    clusters = estimator[2].fit_predict(new)
    score = metrics.silhouette_score(new, clusters)
    return score

def ch_criterion(estimator, X):

    """Takes in pipeline estimator and dataset as arguments.
    
       Function makes predictions of dataset using estimator and computes calinski-harabasz criterion.
       
       Returns computed calinski-harabasz criterion."""
 
    estimator[0].fit(X)
    new = estimator[0].transform(X)
    estimator[1].fit(new)
    new = estimator[1].transform(new)
    clusters = estimator[2].fit_predict(new)
    score = metrics.calinski_harabasz_score(new, clusters)
    return score


#drop unnecessary columns from dataset
X = all_individuals_with_targets.drop(['Name', 'Team', 'Salary', 'target'], axis = 1)
X = X.drop(droplist, axis = 1)

#set random state
random_state = 42

#set cv to zero
cv = [(slice(None), slice(None))]

#create multi-scoring dictionary for gridsearch
scoring = {'Silhouette Score': silhouette_score, 'CH Criterion': ch_criterion}

#create parameters dictionary for gridsearch
params = {'pca__n_components': list(range(2,14)),
           'clust__n_components': list(range(15,56)),
           'clust__covariance_type': ['full', 'tied', 'diag', 'spherical'],
           'clust__init_params': ['kmeans', 'random'],
           'clust__max_iter': [50, 75, 100]}
         

#create empty lists to keep best pipelines and best clfs
pipelinesrl = []
best_clustalgsrl = []
    
#set pipeline with list of standard scalar, pca and cluster algorithm
for scaler in [StandardScaler(), MinMaxScaler()]:
    pipe = Pipeline([('ss', scaler),
                     ('pca', PCA(random_state=random_state)),
                     ('clust', GaussianMixture(random_state=random_state))])

    #conduct gridsearch over parameter variable running on all backend workers
    gs = GridSearchCV(estimator=pipe,
                  param_grid=params,
                  scoring=scoring,
                  refit='Silhouette Score',
                  cv=cv,
                  verbose=1,
                  n_jobs=-1)

    #fit model
    gs.fit(X)
    #save results
    results = gs.cv_results_
    #append best estimator
    best_clustalgsrl.append(gs.best_estimator_)
    #append pipeline
    pipelinesrl.append(pipe)
    #print cluster algorithm and its name
    print('GaussianMixture', 'with', scaler.__class__.__name__)
    #print params of best model
    print('Best params:', gs.best_params_)
    #print silhouette score
    print('Silhouette Score:', silhouette_score(gs.best_estimator_, X))
    #print C-H Criterion
    print('C-H Criterion:', ch_criterion(gs.best_estimator_, X))

    #plot best estimators
    fig, (ax1, ax2) = plt.subplots(2, figsize=(10, 10))
    #set title
    fig.suptitle("GridSearchCV Evaluation Scores by Number of Clusters",
              fontsize=16)
    #set subplot x and y labels
    ax1.set_xlabel("n_clusters")
    ax1.set_ylabel("Silhouette Score")
    #set subplot x and y labels
    ax2.set_xlabel("n_clusters")
    ax2.set_ylabel("C-H Criterion")

    #find index of best silhouette score for the gridsearch
    best_index = np.nonzero(results['rank_test_Silhouette Score'] == 1)[0][0]

    #if else statements to find number of clusters in best estimator
    num_of_clusters = results['param_clust__n_components'][best_index]
    div = 12

    #for loop iterating through a range from 15 to 56, exclusive, to create lists for x-axis and y-axis
    #where x-axis is number of clusters and y-axis is corresponding silhouette score
    x_axis = []
    y_axis_ss = []
    for x in range(15,56):
        difference = num_of_clusters - x
        x_axis.append(x)
        y_axis_ss.append(results['mean_test_Silhouette Score'][best_index - int(difference * div)])

    #for loop iterating through a range from 15 to 56, exclusive, to create list for y-axis
    #where y-axis is corresponding c-h score
    y_axis_ch = []
    for x in range(15,56):
        difference = num_of_clusters - x
        y_axis_ch.append(results['mean_test_CH Criterion'][best_index - int(difference * div)])

    #
    ax1.set_ylim(min(y_axis_ss) - .05, max(y_axis_ss) + .05)
    ax1.plot(x_axis, y_axis_ss, linestyle = '--', color='g', label='Silhouette Score')
    best_score_ss = results['mean_test_Silhouette Score'][best_index]
    # Plot a dotted vertical line at the best score for that scorer marked by x
    ax1.plot([x_axis[num_of_clusters - 15],] * 2, [0, best_score_ss],
            linestyle='-.', color='k', marker='x', markeredgewidth=3, ms=8)
    # Annotate the best score for that scorer
    ax1.annotate("%0.2f" % best_score_ss, (x_axis[num_of_clusters - 15], best_score_ss + 0.005))

    ax2.set_ylim(min(y_axis_ch) - 10, max(y_axis_ch) + 10)
    ax2.plot(x_axis, y_axis_ch, linestyle = '--', color='g', label='CH Criterion')
    best_score_ch = results['mean_test_CH Criterion'][best_index]
    # Plot a dotted vertical line at the best score for that scorer marked by x
    ax2.plot([x_axis[num_of_clusters - 15],] * 2, [0, best_score_ch],
            linestyle='-.', color='k', marker='x', markeredgewidth=3, ms=8)
    # Annotate the best score for that scorer
    ax2.annotate("%0.2f" % best_score_ch, (x_axis[num_of_clusters - 15], best_score_ch + 0.005))

    ax1.legend(loc="best")
    ax1.grid(b = True, which = 'major')
    ax1.grid(b = True, which = 'minor')
    ax2.legend(loc="best")
    ax2.grid(b = True, which = 'major')
    ax2.grid(b = True, which = 'minor')
    plt.show()
```
